```python
# Why distilgpt2?? --> Small, Fast, enough for  assignment purpose.
from transformers import pipeline

qa_chatbot = pipeline(
    task = 'text-generation',
    model = 'distilgpt2'
)
```

```python
def ask_bot(question):
    prompt = f"Question: {question}\nAnswer:"

    response = qa_chatbot(
        prompt,
        max_length=80,
        do_sample=False,
        repetition_penalty=1.5,
        eos_token_id=50256
    )

    # Remove the prompt part
    return response[0]["generated_text"].split("Answer:")[-1].strip()
```

```python
while True:
    q = input("User: ")

    if q.lower() in ["exit", "quit", "stop"]:
        print("Bot: Goodbye!")
        break

    print("Bot:", ask_bot(q))

#turns = 3

#for _ in range(turns):
#    q = input("User: ")
#    print("Bot:", ask_bot(q))

```
